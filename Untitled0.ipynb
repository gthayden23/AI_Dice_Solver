```python
class CubeBalanceEnv(gym.Env):
    def __init__(self):
        self.model = mujoco.MjModel.from_xml_string(free_body_MJCF)
        self.data = mujoco.MjData(self.model)
        self.landing_counts = {'top_bottom': 1, 'sides': 1}  # init to 1 to avoid division by zero
        self.render_counter = 0
        self.viewer = None

        self.action_space = gym.spaces.Box(low=.01, high=0.2, shape=(2,), dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=0, high=1.0, shape=(1,), dtype=np.float32)
        self.recent_landings = deque(maxlen=100)

        self.max_steps = 100
        self.step_count = 0

        self.current_width = 0.06
        self.current_height = 0.06

    def reset(self):
        mujoco.mj_resetData(self.model, self.data)
        self._randomize_initial_orientation()
        self.step_count = 0
        return self._get_obs()
        self.landing_counts = {'top_bottom': 1, 'sides': 1}

    def step(self, action):
        target_height, target_width = action
        alpha = .6  # smoothing factor (0 = no change, 1 = instant change)
        self.current_width = (1 - alpha) * self.current_width + alpha * target_width
        self.current_height = (1 - alpha) * self.current_height + alpha * target_height

        self._update_cube_geometry(self.current_width, self.current_height)

        # Simulate a short drop
        for _ in range(100):  # fixed number of sim steps
            mujoco.mj_step(self.model, self.data)

        z_alignment = self._get_z_alignment()
        observation = np.array([z_alignment], dtype=np.float32)
        reward = self._compute_reward(z_alignment)
        observation = int(abs(z_alignment) > 0.9)
        done = self.data.time >= duration  # Or step counter-based
        return observation, reward, done, {}

    def _compute_reward(self, z_alignment):
      is_tb = abs(z_alignment) > 0.9
      self.recent_landings.append('tb' if is_tb else 's')

      tb = self.recent_landings.count('tb')
      s = self.recent_landings.count('s')
      total = tb + s

      #shaping
      freq_tb = tb / total
      freq_s = s / total

      shaping = 1.0 - (freq_tb if is_tb else freq_s)
      base = 0.2  #small reward

      return base + 0.8 * shaping




    def _update_cube_geometry(self, width, height):
        # Update face sizes
        self.model.geom_size[2] = np.array([0.001, width, height])
        self.model.geom_size[3] = np.array([0.001, width, height])
        self.model.geom_size[4] = np.array([width, 0.001, height])
        self.model.geom_size[5] = np.array([width, 0.001, height])
        self.model.geom_size[6] = np.array([width, width, 0.001])
        self.model.geom_size[7] = np.array([width, width, 0.001])

        self.model.geom_pos[2] = np.array([ width, 0, 0.6])
        self.model.geom_pos[3] = np.array([-width, 0, 0.6])
        self.model.geom_pos[4] = np.array([0,  width, 0.6])
        self.model.geom_pos[5] = np.array([0, -width, 0.6])
        self.model.geom_pos[6] = np.array([0, 0, 0.6 + height])
        self.model.geom_pos[7] = np.array([0, 0, 0.6 - height])

    def _get_z_alignment(self):
        cube_body_id = self.model.body('cube').id
        xmat = self.data.xmat[cube_body_id].reshape(3, 3)
        world_z = xmat @ np.array([0, 0, 1])
        return world_z[2]

    def _get_obs(self):
        return np.array([self._get_z_alignment()], dtype=np.float32)

    def _randomize_initial_orientation(self):
        self.data.qvel[3:6] = 3*np.random.randn(3)

    def render(self, mode = "fixed"):
        self.render_counter += 1
        if self.render_counter % 1000 == 0:
            print(f"[Render] Showing model at step {self.render_counter}")
            mujoco.viewer.launch_passive(self.model, self.data)  # Opens new window every 100 steps

```

```python
def make_flat_sphere(radius, depth):
    flat_depth = radius* depth/100
    flat_depth *= 2
    sphere = trimesh.creation.icosphere(subdivisions=4, radius=radius)

    # plane
    plane_origin = np.array([0, 0, -radius + flat_depth])  #height
    plane_normal = np.array([0, 0, 1])  #horizontally
                                                                                                                                #5:0 4:1 3:2 2:3 1:4 0:5
                                                                                                                                #5:0 4:.5 3:1
    cylinder = trimesh.creation.cylinder(radius=pow(pow(radius, 2)-pow(radius-flat_depth, 2), .5), height=(radius-flat_depth + ((radius*.5 - flat_depth*.5) + radius*.5 - .5*flat_depth)))
    rot_matrix = transformations.rotation_matrix(3.14159, [0, 1, 0])

    cylinder.apply_transform(rot_matrix)

    # Slice
    sliced = sphere.slice_plane(plane_origin, plane_normal)
    sliced.fill_holes()
    filled_sliced = trimesh.util.concatenate([cylinder, sliced])
    filled_sliced.fill_holes()

    os.chdir("./saved_models")
    # Export to STL for MuJoCo
    mesh.export('flat_sphere.stl')
    os.chdir("\\")
    return filled_sliced

mesh = make_flat_sphere(.1, 25)
```
